````python
import numpy as np
import matplotlib.pyplot as plt

# Generate a list of 100 random numbers using NumPy's normal distribution function
# with a mean of 0 and a standard deviation of 1
random_numbers = np.random.normal(loc=0, scale=1, size=100)

# Plot the random numbers with green lines, using asterisks as markers
# with blue marker faces, a markersize of 10, and a dashed line style
plt.plot(random_numbers, color='green', marker='*', markerfacecolor='blue', markersize=10, linestyle='--', markeredgewidth=2, markeredgecolor='black')

# Plot the same random numbers with red circles as markers, using no line style
# with red marker faces, a markersize of 7, and no marker edge width or color
plt.plot(random_numbers, color='green', marker='o', markerfacecolor='red', markersize=7, linestyle='none', markeredgewidth=0)

# Add a title and labels to the x and y axes
plt.title('Random Numbers')
plt.xlabel('Index')
plt.ylabel('Value')

# Save the plot as a PNG file with the filename 'random_numbers.png'
plt.savefig('random_numbers.png')

# Show the plot on the screen
plt.show()
````

we used NumPy to generate a list of 100 random numbers from a normal distribution with mean 0 and standard deviation 1. We specified the mean and standard deviation with the loc and scale parameters of the `np.random.normal()` function, and we set the size of the list with the size parameter.

We then used Matplotlib to create a plot of the random numbers. We customized the plot by specifying different marker styles and colors for each set of data points.
`plt.plot(random_numbers, color='green', marker='*', markerfacecolor='blue', markersize=10, linestyle='--', markeredgewidth=2, markeredgecolor='black')`
The first set of data points are plotted with green dashed lines and blue star markers with black borders.
`plt.plot(random_numbers, color='green', marker='o', markerfacecolor='red', markersize=7, linestyle='none', markeredgewidth=0)`
The second set of data points are plotted with no lines and red circle markers with blue centers. We also specified the size of the markers with the markersize parameter and the width of the marker edges with the markeredgewidth parameter.

plt.xlabel('Index')
plt.ylabel('Value')```
We added a title to the plot with the title() function, and labeled the x and y axis with the `xlabel()` and `ylabel()` functions. Finally, `plt.savefig('random_numbers.png')` we saved the plot as a PNG file with the `savefig()` function and displayed it with the `show()` function.
